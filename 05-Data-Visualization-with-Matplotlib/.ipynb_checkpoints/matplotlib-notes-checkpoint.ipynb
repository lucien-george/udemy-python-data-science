# Matplotlib Notes

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, c='r')

plt.subplot(1, 2, 2)
plt.plot(y, x, c='b')
```

```python
# OOP method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # [left_offset, bottom_offset, width, height]
axes1.plot(x, y)
axes1.set_title('Larger plot')
axes2.plot(y, x)
axes2.set_title('Smaller plot')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))

ax1.plot(x, y)
ax1.set_title('first plot')
ax2.plot(y, x)
ax2.set_title('second plot')
plt.tight_layout()
```

## Figure size and DPI

```python
fig = plt.figure(figsize=(15, 8))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, label="X squared")
ax.plot(x, x**3, label="X cubed")
ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')
ax.legend(loc='best')
```

## Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='r', linewidth=1, alpha=0.7, linestyle='--', marker='o',
        markersize=15, markerfacecolor='blue', markeredgewidth='3', markeredgecolor='yellow')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
